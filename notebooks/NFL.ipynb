```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

print(Path.cwd())
#received error code csv not found, use this to see where python is pulling from and noticed i saved it the wrong folder
```

```python
df = pd.read_csv("team_stats_2003_2023.csv")
print(df.head())
```

```python
df.columns.tolist()
```

```python
#creating histograms for relevant columns/features
columns_to_plot = [
    "pass_yds",
    "pass_td",
    "pass_att",
    "pass_cmp",
]
plt.figure(figsize=(12, 8))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(3, 2, i)
    plt.hist(df[col], bins=15, color="red", edgecolor="black")
    plt.title(col)
    plt.xlabel(col)
plt.tight_layout()
plt.show()
```
